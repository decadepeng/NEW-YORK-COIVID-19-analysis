**NY CA Covid-19 Analysis**

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
NY_county_wise = pd.read_csv("NY_County.csv")
NY_county_wise
```

```python
NY_county_wise_tidy = NY_county_wise[['cases', 'deaths', 'fully_vaccinated']]
NY_county_wise_tidy.head()
```

```python
from sklearn.cluster import KMeans
NY_k_means = KMeans(n_clusters=3)
NY_kmeans = NY_k_means.fit(NY_county_wise_tidy)
NY_kmeans.cluster_centers_

NY_county_wise_tidy['cluster'] = NY_kmeans.labels_
NY_county_wise_tidy
```

```python
NY_county_wise_tidy["county"] = NY_county_wise['county']
NY_county_wise_tidy
```

```python
NY_county_wise_3D_plot = px.scatter_3d(
    data_frame = NY_county_wise_tidy,
    x = 'cases',
    y = 'deaths',
    z = 'fully_vaccinated',
    template = 'ggplot2',
    title = 'NY Covid-19 Countywise Analysis',
    color = 'cluster',
    hover_name = 'county',

)

NY_county_wise_3D_plot
```

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

**NYC Daywise Anlysis**


**Ingestion**


```python
NYC_day_wise = pd.read_csv("NYC_DAYWISE.csv")
NYC_day_wise
```

**DATA CLEAN**

```python
NYC_day_wise_tidy = NYC_day_wise[['CASE_COUNT', 'HOSPITALIZED_COUNT', 'DEATH_COUNT']]
NYC_day_wise_tidy.head()
```

# 新段落

**Clustering**

```python
from sklearn.cluster import KMeans
NYC_k_means = KMeans(n_clusters=2)
NYC_kmeans = NYC_k_means.fit(NYC_day_wise_tidy)
NYC_kmeans.cluster_centers_
```

```python
NYC_day_wise_tidy['cluster'] = NYC_kmeans.labels_
NYC_day_wise_tidy
```

```python
NYC_day_wise_tidy["date"] = NYC_day_wise['date_of_interest']
NYC_day_wise_tidy
```

**Plot the 3D graph**

```python
NYC_day_wise_3D_plot = px.scatter_3d(
    data_frame = NYC_day_wise_tidy,
    x = 'CASE_COUNT',
    y = 'HOSPITALIZED_COUNT',
    z = 'DEATH_COUNT',
    template = 'ggplot2',
    title = 'NYC Covid-19 Daywise Analysis',
    color = 'cluster',
    hover_name = 'date',

)

NYC_day_wise_3D_plot
```
